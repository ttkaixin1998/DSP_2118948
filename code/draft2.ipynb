```python
all_list=[]
for i in range(len(dpa)):
    dpa[i].lower()
    for s in ',.\n ':
        dpa[i]=dpa[i].replace(s,' ')
    single_list=dpa[i].split()
    single_counts={}
    for word in single_list:
        if word in single_counts: 
            single_counts[word] = single_counts[word] + 1 
        else: 
            single_counts[word] = 1
    sort=sorted(single_counts.items(), key=lambda item:item[1],reverse=True)
    print('dpa'+'['+str(i)+']')
    print(sort[:10])
    all_list.append(sort)
```

```python
all_freq=[]
for i in range(len(dpa)):
    dpa[i].lower()
    for s in ',.\n ':
        dpa[i]=dpa[i].replace(s,' ')
    noun_list=dpa[i].split()
    noun_counts={}
    noun=['NN','NNS','NNP','NNPS']
    for word, tag in nltk.pos_tag(noun_list):
        if word in noun_counts and tag in noun: 
            noun_counts[word] = noun_counts[word] + 1 
        else: 
            noun_counts[word] = 1
    sort=sorted(noun_counts.items(), key=lambda item:item[1],reverse=True)
    print('dpa'+'['+str(i)+']')
    print(sort[:5])
    all_freq.append(sort[0])
```

```python
all_freq_dict={}
for word in all_freq:
    if word[0] in all_freq_dict: 
        all_freq_dict[word[0]] = all_freq_dict[word[0]] + word[1] 
    else: 
        all_freq_dict[word[0]] = word[1]
```

```python
all_location_list = []
article=xiaomi
test_article=google
for doc in article:
    doc_list = [word for word in jieba.cut_for_search(doc)]
    all_location_list.append(doc_list)
    
# 制作语料库,获取词袋
dictionary = corpora.Dictionary(all_location_list)
corpus = [dictionary.doc2bow(doc) for doc in all_location_list]
# 使用TF-IDF模型对语料库建模
tfidf = models.TfidfModel(corpus)

#特征数
featureNUM = len(dictionary.token2id.keys())
#通过TfIdf对整个语料库进行转换并将其编入索引，以准备相似性查询
index = similarities.SparseMatrixSimilarity(tfidf[corpus],num_features=featureNUM)
#稀疏向量.dictionary.doc2bow(doc)是把文档doc变成一个稀疏向量，[(0, 1), (1, 1)]，表明id为0,1的词汇出现了1次，至于其他词汇，没有出现。

abc=[]
for i in range(len(test_article)):
    test_article_slice= google[i]
    test_article_list = [word for word in jieba.cut_for_search(test_article_slice)]
    # 测试文档的二元组向量转换
    new_vec = dictionary.doc2bow(test_article_list)
    # 获取测试文档中，每个词的TF-IDF值
    tfidf[new_vec]
    #计算向量相似度
    simi = index[tfidf[new_vec]]
    abc.append(['Similarity for each page of xiaomi and google'+str(i),simi.max()])
```
